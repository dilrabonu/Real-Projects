```python
import json

# Step 1: Load the JSON file
with open("/content/data_description_rating.json", "r", encoding="utf-8") as infile:
    movies = json.load(infile)

# Step 2: Write in bulk (NDJSON) format
with open("/content/bulk_movies.json", "w", encoding="utf-8") as outfile:
    for movie in movies:
        index_cmd = {"index": {"_index": "movies", "_id": movie["movieId"]}}
        outfile.write(json.dumps(index_cmd) + "\n")
        outfile.write(json.dumps(movie) + "\n")

print("✅ bulk_movies.json created successfully!")

```
